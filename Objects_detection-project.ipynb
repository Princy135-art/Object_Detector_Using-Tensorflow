# Import Libraries


```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow_hub as hub
```

```python
model= hub.load("https://tfhub.dev/tensorflow/ssd_mobilenet_v2/fpnlite_320x320/1")
```

```python

import cv2
image_path= "object_detection/images/room.jpg"
image= cv2.imread(image_path)
print(image)
```

```python

import matplotlib.pyplot as plt
image_rgb= cv2.cvtcolor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
import tensorflow as tf
input_tensor= tf.image.resize(image_rgb, (320,320))
input_tensor= tf.cast(input_tensor, tf.uint8)
input_tensor= input_tensor[tf.newaxis,...]
```

```python
detections= model(input_tensor)
```

```python
print(detections['detection_scores'])
```

```python
image_with_boxes = image.copy()
h, w, _ = image.shape
for i in range(10):
    score = detections['detection_score'][0][i].numpy()
    if score>0.5:
        box = detections['detection_boxes'][0][i].numpy()
        y1, x1, y2, x2 = box
        cv2.rectangle(
        image_with_boxes,
        (int(x1*w), int(y1*h)),
        (int(x2*w), int(y2*h)),
        (0,225,0),
        2
        )
```

```python
plt.imshow(cv2.cvtcolor(image_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
